```python
pd.set_option('display.max_rows', None)

import numpy as np
col_name= []
mean_val=[]
nulls=[]

for x in df.columns:
    try:
        #print(x, "| Mean: ", round(np.mean(df[x]),2), "| null values: ", df[x].isnull().sum())
        mean_val.append(round(np.mean(df[x]),3))
        nulls.append(df[x].isnull().sum())
        col_name.append(x)
    except:
        #print(x)
        col_name.append(x)
        mean_val.append(0)
        nulls.append(df[x].isnull().sum())
    
summary_df = pd.DataFrame({"field":col_name,
                          "Mean Value":mean_val,
                          "Null Values": nulls})

summary_df.sort_values(by=['Null Values'], ascending=False)
```
